# 대박은 항상 소수다

- '거대한 지진은 매우 드물게 일어나지만, 미약한 지진은 자주 발생한다'
- 초대형 도시가 소수 나타나고, 비슷하게 조그마한 도시가 많이 나타난다.
- 웹사이트 트래픽도 승자독식이다. 트래픽이 많은 거대 웹사이트 몇개와 트래픽이 그저그런 수많은 웹사이트가 존재한다.

이는 파레토 법칙이라고 하는 `80-20`원칙과도 연관이 있다. 은행 수익 80%는 20%의 고객으로부터 비롯된다던지, 전세계 부의 80%는 20% 소수에게 집중되어있다던지 하는 것이다. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# np.random.power(a,b)
# a가 클수록 1에 얼마나 가까운 숫자들일지 (편향성)
# b는 size
```

```python
a = 50.
samples = 10000
s = np.random.power(a, samples)
```

```python
count, bins, ignored = plt.hist(s, bins=30)
```

```python
x = np.linspace(0, 1, 100)
```

```python
y = a*x*(a-1.)
```

```python
normed_y = samples*np.diff(bins)[0]*y
```

```python
plt.plot(x, normed_y)
```

```python
from scipy import stats
```

```python
rvs = np.random.power(5, 1000000)
```

```python
rvsp = np.random.pareto(5, 1000000)
```

```python
xx = np.linspace(0,1,100)
```

```python
powpdf = stats.powerlaw.pdf(xx,5)
```

```python
plt.figure()
plt.hist(rvs, bins=50, normed=True)
plt.plot(xx, powpdf, 'r-')
plt.title('np.random.power(5)')
```

```python
plt.figure()
plt.hist(1./(1.+rvsp), bins=50, normed=True)
plt.plot(xx, powpdf, 'r-')
plt.title('inverse of 1 + np.random.pareto(5)')
```

```python
plt.figure()
plt.hist(1./(1.+rvsp), bins=50, normed=True)
plt.plot(xx, powpdf, 'r-')
plt.title('inverse of stats.pareto(5)')
```
